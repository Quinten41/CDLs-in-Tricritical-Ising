# First let's calculate some modular matrices in tricritcal Ising and tricritical Ising squared!

```python
'''
This block sets some parameters for tricritical ising and fixes a fundamental domain for the Kac labels (called Epp)
It also computes the modular matrices after gauging by the Z_2 line eta in the original and doubled theory
'''

# This will be the (p,p'=pp) minimal model
p = 5  # p
pp = 4  # p'
# Various constant parameters of the minimal model
N = 2 * p * pp
num_prim = (p-1)*(pp-1) // 2  # Number of primaries

c = 1 - 6 * (p - pp)**2 / (p * pp) # Central charge

# Index set
####### This needs to be changed for each minimal model ########
Epp = [ (1,1), (1,2), (1,3), (1,4), (2,2), (2,4) ] # By (r,s); this is for tricritical Ising
# The conformal weights
cdim = [ ((p*r-pp*s)**2 - (p-pp)**2) / (2*N) for r,s in Epp ]

# Modular S-matrix
S = matrix( [ [ 4*sqrt( 1/QQ(N) )*(-1)**(1+s*u+r*v)*sin(pi*QQ(p)/QQ(pp)*r*u)*sin(pi*QQ(pp)/QQ(p)*s*v) for r,s in Epp ] for u,v in Epp ] )
S = matrix( mathematica(S).Simplify().sage() )
# Modular T-matrix
T = diagonal_matrix( [ exp(2*pi*I*(h-c/24)) for h in cdim ] )

# We calculate the M matrix (called Msg) of the minimal model squared gauged by eta \otimes eta where eta is the Z_2 Verlinde line.
M = identity_matrix(num_prim)
Mpp = M
Mpm = diagonal_matrix( [ (-1)**(r+1) for r,_ in Epp ] ) 
Mmp = matrix( mathematica( S.H*Mpm*S ).FullSimplify().sage() )
Mmm = matrix( mathematica( T.H*Mmp*T ).FullSimplify().sage() )

Mg = matrix( mathematica((Mpp+Mpm+Mmp+Mmm)/2).FullSimplify().sage() ) # Gauging by eta in the non-squared theory

Msg = matrix( mathematica( (Mpp.tensor_product(Mpp)+Mpm.tensor_product(Mpm)\
                               +Mmp.tensor_product(Mmp)+Mmm.tensor_product(Mmm))/2 ).FullSimplify().sage() )
```

# Now we will compute the $q$ expansions of some partition functions in tricritical Ising (squared).

```python
%%cython
from sage.rings.power_series_ring import PowerSeriesRing
from sage.rings.puiseux_series_ring import PuiseuxSeriesRing
from sage.rings.rational_field import QQ
from sage.rings.integer_ring import ZZ
from sage.rings.imaginary_unit import I
from sage.combinat.partition import number_of_partitions
from sage.functions.other import ceil
from sage.matrix.special import identity_matrix

'''
This code should calculate the characters and partition function of a minimal model (tricirtical ising)!
It can also be used to calculate variants of the partition function useful for calculating a gauged partition function
'''


# Minimal Model Fixed Parameters
# Defined at compile time

# This will be the (p,p'=pp) minimal model
DEF p = 5  # p
DEF pp = 4  # p'
# Various constant parameters of the minimal model
DEF N = 2 * p * pp
DEF num_prim = (p-1)*(pp-1) // 2  # Number of primaries

c = 1 - QQ(6 *(p-pp)**2)/QQ(p * pp) # Central charge


# Index set
####### This needs to be changed for each minimal model ########
cdef list Epp = [(1,1), (1,2), (1,3), (1,4), (2,2), (2,4)]


cdef int isqrt(int n):
    '''
    Helper function to compute the integer square root plus one
    '''

    cdef int x = n
    cdef int y = 1
    while x > y:
        x = (x + y) >> 1
        y = n // x
    return x+1


def chi_int(int n_max, int r, int s):
    '''
    Function to compute the character \chi_{\lambda_{r,s}}(q)*q^{-h_{r,s} + c/12}.
    This only has integer powers and will be computed up to O(q^{n_max}).
    '''

    # Define the power series ring the characters will belong to
    # The output will be an element of this ring
    P = PowerSeriesRing(ZZ, 'q', sparse = False, default_prec = n_max)

    # Define an effective maximum for n to loop to below so we don't add more terms than necessary
    cdef int eff_n_max = ( p*r + pp*s + isqrt( (p*r-pp*s)**2 + 2*n_max*N ) ) // N + 1

    # Return the difference of the two K-functions times eta^{-1} (but with only integer powers)
    return ( P({ n*(n*p*pp+p*r-pp*s) : 1 for n in range(-eff_n_max, eff_n_max) }, n_max) +\
    P({ (n*p + s)*(n*pp + r) : -1 for n in range(-eff_n_max, eff_n_max) }, n_max) ) *\
    P([ number_of_partitions(n) for n in range(n_max+1) ])

    
def Z(int n_max, list part_funcs = [[ identity_matrix(ZZ, num_prim), 1, 1, 1, 1 ]]):
    '''
    Function to compute the partition function!
    
    If only n_max is given it will just compute the partition function of the specified minimal model up to O(q^{n_max}).
    
    If part_funcs is given the notation should be as follows: part_funcs = [L_1 , ... , L_m] (m can be anything). Each L_i should
    take the form L_i = [M, fac, expo, ov_fac, pow] where M is a square, integer matrix whose size is the same as the number of 
    primaries, fac is a non-negative integer, expo is a positive rational number, ov_fac is an integer, and pow is a non-negative 
    integer. What is returned is th sum of all ov_fac*Z_M(fac*q^expo)^pow expanded around q=0 up to O(q^{n_max}).
    '''

    # Define the power series ring the partition function will belong to
    # The output will be an element of this ring
    P = PuiseuxSeriesRing(ZZ, 'q', sparse = True, default_prec = n_max)
    
    # Initialise the output series as the zero element
    Z_out = P(0)

    # Find up to what order we have to compute the characters
    min_exp = min( part_funcs, key = lambda x:x[2] )[2]
    cdef int eff_n_max = ceil(n_max/min_exp) + 1

    # For faster computation a power series ring with only integer powers is defined
    PZ = PowerSeriesRing(ZZ, 'q', sparse = False, default_prec = eff_n_max)

    # Finds the products of all pairs of characters
    cdef int i,j
    cdef list all_chi = [ 0 for _ in range(num_prim) ]
    cdef list chi_prod = [ [0 for _ in range(num_prim)] for _ in range(num_prim) ]
    for i in range(num_prim):
        all_chi[i] = PZ(chi_int(eff_n_max, Epp[i][0], Epp[i][1]), eff_n_max)
        chi_prod[i][i] = all_chi[i]**2
        for j in range(i):
            chi_prod[i][j] = all_chi[i]*all_chi[j]

    # Loops over all the functions in part_funcs we are summing over
    cdef int rp, sp, r, s, n, M_eff
    cdef int nexpo, dexpo, nshift, dshift, deg, p1, p2, pow
    for [ M, fac, expo, ov_fac, pow ] in part_funcs:
        eff_n_max = ceil(n_max/expo) + 1
        expo = QQ(expo)
        nexpo = expo.numerator()
        dexpo = expo.denominator()
        fac = ZZ(fac)
        Z_sub = P(0)
        for i in range(num_prim):
            r = Epp[i][0]
            s = Epp[i][1]
            if M[i,i] != 0:
                this_chi = chi_prod[i][i]
                shift = QQ( (p * r - pp * s)**2 - (p-pp)**2 ) / QQ( N ) - c/12
                nshift = shift.numerator()
                dshift = shift.denominator()
                deg = dexpo*dshift
                p1 = nexpo*dshift
                p2 = nshift*nexpo
                Z_sub += P( {p1*n+p2 : this_chi[n] * fac**n * M[i,i] for n in range(eff_n_max)}, deg, n_max )
            for j in range(i):
                M_eff = M[i,j] + M[j,i]
                if M_eff != 0:
                    this_chi = chi_prod[i][j]
                    rp = Epp[j][0]
                    sp = Epp[j][1]
                    shift = (QQ((p * r - pp * s)**2 - (p-pp)**2) + QQ((p * rp - pp * sp)**2 - (p-pp)**2)) / QQ(2*N) - c/12
                    nshift = shift.numerator()
                    dshift = shift.denominator()
                    deg = dexpo*dshift
                    p1 = nexpo*dshift
                    p2 = nshift*nexpo
                    Z_sub += P( {p1*n+p2 : this_chi[n] * fac**n * M_eff for n in range(eff_n_max)}, deg, n_max )
                    
        Z_out += P(ov_fac) * Z_sub**pow
    
    return Z_out
```

# Compute the characters and partition function of the coset model $SU(2)_8 / U(1)$.

```python
%%cython
from sage.rings.power_series_ring import PowerSeriesRing
from sage.rings.puiseux_series_ring import PuiseuxSeriesRing
from sage.rings.rational_field import QQ
from sage.rings.integer_ring import ZZ
from sage.combinat.partition import number_of_partitions
from sage.functions.other import ceil

# Define some compile time parameters
DEF k = 8 # level
DEF num_prim = k*(k+1) // 2 # number of primary fields (primary with respect to the chiral algebra, of course)
c = QQ(2*(k-1))/QQ((k+2)) # central charge

# Define a fundamental domain for l,m to live in; our convention is that l,m are integers (twice the physics convention).
cdef int m_max(int l):
    if l <= k/2:
        return l
    else:
        return l-2
cdef int l,m
cdef list fund = [(l,m) for l in range(k+1) for m in range(-m_max(l), m_max(l)+1, 2)]
# This is just a weird degeneracy not covered by the above
if k % 2 == 0:
    fund.remove( (k//2, -k//2) )

# I don't think this can ever happen, but if it does something went very wrong.
if num_prim != len(fund):
    raise RuntimeError('The number of indices in the fundamental domain does not match the number of primaries.')

# The conformal weights
cdef list weights = [ QQ(l*(l+2)) / QQ(4*(k+2)) - QQ(m**2) / QQ(4*k) for (l,m) in fund ]

cdef int isqrt(int n):
    '''
    Helper function to compute the integer square root plus one
    '''
    cdef int x = n
    cdef int y = 1
    while x > y:
        x = (x + y) >> 1
        y = n // x
    return x+1

cdef int r_max1(int n_max, int s, int m, int l):
    print((-l - m - 18*s + isqrt(l**2 + 2*(l + 1)*m + m**2 + 8*(4*l + 5*m + 4)*s + 320*s**2 + 2*l + 8*n_max + 1) - 1)//2 + 1)
    return (-l - m - 18*s + isqrt(l**2 + 2*(l + 1)*m + m**2 + 8*(4*l + 5*m + 4)*s + 320*s**2 + 2*l + 8*n_max + 1) - 1)//2 + 1

cdef int r_max2(int n_max, int s, int m, int l):
    return (l + m - 18*s + isqrt(l^2 + 2*(l - 19)*m + m^2 - 8*(4*l + 5*m - 76)*s + 320*s^2 - 30*l + 8*n_max + 289) - 19)//2 + 1

def chi_int_paraferm(int n_max, int l, int m):
    '''
    This computes the charcters of the coset model SU(2)_8/U(1) 
    Only the integer powers are computed; the q^{h-c/24} must be added afterwards
    The character is computed up to order O(q^{n_max})
    '''

    # Define the power series ring the characters will belong to
    # The output will be an element of this ring
    P = PowerSeriesRing(ZZ, 'q', sparse = False, default_prec = n_max)

    cdef int r,s,eff_n_max = 2*n_max
    P1 = sum( P({ ( r*(r+1+l+m) + s*(s+1+l-m) )//2 + r*s*(k+1) : (-1)**(r+s)\
            for r in range( eff_n_max ) }, eff_n_max) for s in range(eff_n_max) )
    P2 = sum( P({ ( r*(r+1-l-m) + s*(s+1-l+m) )//2 + (k+1)*(r+s+1+r*s) - l : (-1)**(r+s+1)\
            for r in range( eff_n_max ) }, eff_n_max) for s in range(eff_n_max) )
    eta_inv_sq = P([ number_of_partitions(r) for r in range(eff_n_max) ], eff_n_max)**2

    return (P1 + P2) * eta_inv_sq


def Zc(int n_max):
    '''
    This functions computes the partition function of the coset SU(2)_8/U(1) up to order O(q^{n_max})
    '''
    
    P = PuiseuxSeriesRing(ZZ, 'q', sparse = True, default_prec = n_max)
    q = P.gen()
    Z_out = P(0)
    PZ = PowerSeriesRing(ZZ, 'q', sparse = False, default_prec = n_max)
    cdef int i,l,m
    for i in range(num_prim):
        (l,m) = fund[i]
        h = weights[i]
        Z_out += q**(2*h - c/12) * P(PZ(chi_int_paraferm(n_max, l, m), n_max+2)**2)

    return Z_out.add_bigoh(n_max)
```

# Compute the decomposition of characters in $SU(2)_8 / U(1)$ into twisted characters of tricritical Ising.

```python
%%time
%%cython
from sage.rings.power_series_ring import PowerSeriesRing
from sage.rings.puiseux_series_ring import PuiseuxSeriesRing
from sage.modular.etaproducts import qexp_eta
from sage.rings.rational_field import QQ
from sage.rings.qqbar import QQbar
from sage.rings.integer_ring import ZZ
from sage.combinat.partition import number_of_partitions
from sage.functions.other import ceil
from sage.matrix.special import identity_matrix
from sage.all import I,pi,exp,matrix,vector,show,sqrt,sin,diagonal_matrix,round,SR,mathematica,pretty_print
import itertools as itr

'''
This code should calculate the cyclic orbifold characters of tricritical Ising squared and then the 
decomposition of the characters of SU(2)_8/U(1) characters into the cyclic orbifold characters of tricritical Ising squared.
'''


# Minimal Model Fixed Parameters
# Define these at compile time

# This will be the (p,p'=pp) minimal model
DEF p = 5  # p
DEF pp = 4  # p'
# Various constant parameters of the minimal model
DEF N = 2 * p * pp
DEF num_prim = (p-1)*(pp-1) // 2  # Number of primaries

c = 1 - QQ(6 *(p-pp)**2)/QQ(p * pp) # Central charge

# Index set
####### This needs to be changed for each minimal model ########
cdef list Epp = [ (1,1), (1,2), (1,3), (1,4), (2,2), (2,4) ] # By (r,s)
# The conformal weights
cdef list cdim = [ (QQ(p*r-pp*s)**2 - QQ(p-pp)**2) / QQ(2*N) for r,s in Epp ]

# Modular S-matrix
S = matrix( [ [ 4*sqrt( 1/QQ(N) )*(-1)**(1+s*u+r*v)*sin(pi*QQ(p)/QQ(pp)*r*u)*sin(pi*QQ(pp)/QQ(p)*s*v) for r,s in Epp ] for u,v in Epp ] )
S = matrix( mathematica(S).Simplify().sage() )
# Modular T-matrix
T = diagonal_matrix([exp(2*pi*I*(h-c/24)) for h in cdim])
Tsq = diagonal_matrix([exp(4*pi*I*(h-c/24)) for h in cdim]) # Modular T-matrix squared

# The precision of the calculation
# Each series will be computed up to this order in q
DEF n_max = 50


cdef int isqrt(int n):
    '''
    Helper function to compute the integer square root plus one
    '''
    cdef int x = n
    cdef int y = 1
    while x > y:
        x = (x + y) >> 1
        y = n // x
    return x+1


def chi_int(int this_n_max, int r, int s):
    '''
    Function to compute the character \chi_{\lambda_{r,s}}(q)*q^{-h_{r,s} + c/12}.
    This only has integer powers and will be computed up to O(q^{n_max}).
    '''

    # Define the power series ring the characters will belong to
    # The output will be an element of this ring
    P = PowerSeriesRing(ZZ, 'q', sparse = False, default_prec = this_n_max)

    # Define an effective maximum for n to loop to below so we don't add more terms than necessary
    cdef int eff_n_max = ( p*r + pp*s + isqrt( (p*r-pp*s)**2 + 2*this_n_max*N ) ) // N + 1

    # Return the difference of the two K-functions times eta^{-1} (but with only integer powers)
    return ( P({ n*(n*p*pp+p*r-pp*s) : 1 for n in range(-eff_n_max, eff_n_max) }, this_n_max) +\
    P({ (n*p + s)*(n*pp + r) : -1 for n in range(-eff_n_max, eff_n_max) }, this_n_max) ) *\
    P([ number_of_partitions(n) for n in range(this_n_max+1) ])


'''
The following code actually computes the characters with respect to the invariant algebra
'''

# For faster computation a power series ring with only integer powers is defined
PZ = PowerSeriesRing(ZZ, 'q', sparse = False, default_prec = n_max)

# Power Series Ring the characters will live in
P = PuiseuxSeriesRing(QQ, 'q', sparse = True, default_prec = n_max)
q = P.gen()


def shift_series(ser, shift, prec = n_max):
    '''
    Helper function to shift the powers of a series by some fixed amount called 'shift'
    '''
    
    nshift = shift.numerator()
    dshift = shift.denominator()
    try:
        ram = QQ(ser.ramification_index())
    except AttributeError:
        return P({ZZ(k*dshift + nshift) : ser[k] for k in ser.exponents()}, dshift, prec)
    else:
        return P({ZZ(ram*rat*dshift+ram*nshift) : ser[rat] for rat in ser.exponents()}, dshift*ram, prec)


def twist_char_update(character, index):
    '''
    Helper function to update the dictionary twist_char with a newly calculated twist character
    '''
    
    wght = character.valuation() + c/12
    if wght in twist_char:
        twist_char[wght][index] = character
    else:
        twist_char[wght] = { index : character }

# This code actually computes the characters with respect to the invariant algebra
# These include the characters twisted by the exchange defect hence the name 'twist_char' for the dictionary storing them
# Because characters with respect to the invariant algebra is a mouthfull, 
# I abuse terminology and call all these characters twist characters in the comments
cdef int i,j,k
cdef list all_chi = [ 0 for _ in range(num_prim) ]
cdef dict twist_char = {}
for i in range(num_prim):
    all_chi[i] = PZ(chi_int(2*n_max+1, Epp[i][0], Epp[i][1]), 2*n_max+1)
    this_chi = PZ (all_chi[i] )

    # 'Diagaonal' Untwisted twist characters
    twist_char_update( shift_series( ( this_chi**2 + PZ({2*k:this_chi[k] for k in range(n_max)}) )/2, 2*cdim[i]-c/12, n_max ), (i,'p') )
    twist_char_update( shift_series(  ( this_chi**2 - PZ({2*k:this_chi[k] for k in range(n_max)}) )/2, 2*cdim[i]-c/12, n_max ), (i,'m') )

    # Twisted twist characters
    twist_char_update( shift_series( P([this_chi[k]*(1-k%2) for k in range(2*n_max)], QQ(2)), cdim[i]/2-c/48, n_max ), (i,'tp') )
    twist_char_update( shift_series( P([this_chi[k]*(k%2) for k in range(2*n_max)], QQ(2)), cdim[i]/2-c/48, n_max ), (i,'tm') )

    # Untwisted twisted characters 
    for j in range(i):
        twist_char_update( shift_series(this_chi*PZ(all_chi[j]), cdim[i]+cdim[j]-c/12, n_max), (i,j) )

cdef set conf_w = set( twist_char.keys() )

# Now we define an index list for the twisted characters to go between the (i,something) notation and a single index
twist_index = [ (i,j) for i in range(num_prim) for j in range(i) ] + [ (i,'p') for i in range(num_prim) ]\
+ [ (i,'m') for i in range(num_prim) ] + [ (i,'tp') for i in range(num_prim) ] + [ (i,'tm') for i in range(num_prim) ]
# Define a list containing the twist_char conformal weights in the order of the above list
cdef dict index_to_weight = { index : wght for wght in conf_w for index in twist_char[wght].keys() } # This is a 'helper' dictionary
cdef list weight_index = [ index_to_weight[index] for index in twist_index ] # This is the actual list
# Number of twisted characters
num_twist_char = len(twist_index)

print([QQ(k)-QQ(c)/QQ(12) for k in twist_char.keys()])
print(conf_w)
print([twist_char[wght][index].valuation() for wght in twist_char for index in twist_char[wght]])

'''
The following code will generate the S-matrix for the twisted characters.
It was somputed by hand in terms of the S-matrix for Tricritical Ising
And this code just copies this computation
'''
S_twist = matrix(SR, [[0 for _ in range(num_twist_char)] for _ in range(num_twist_char)])
Comb_Mat = matrix( mathematica(S*Tsq*S).Simplify().sage() )
cdef int num_pair = round( num_prim*(num_prim-1)/2 )
cdef dict ind_sign = {'p':1, 'm':-1, 'tp':1, 'tm':-1}


def better_simplify(expr):
    '''
    Helper function that first trys to use mathematica's FullSimplify, then switches to giac's simplify if this fails
    '''
    try:
        return mathematica(expr).FullSimplify().sage()
    except:
        return expr.simplify(algorithm = 'giac')

for i in range(num_pair):
    for j in range(num_pair):
        S_twist[i,j] = better_simplify( S[twist_index[i][0],twist_index[j][0]] * S[twist_index[i][1],twist_index[j][1]]\
        + S[twist_index[i][1],twist_index[j][0]] * S[twist_index[i][0],twist_index[j][1]] )
    for j in range(num_pair, num_pair + 2*num_prim):
        S_twist[i,j] = better_simplify( S[twist_index[i][0],twist_index[j][0]]*S[twist_index[i][1],twist_index[j][0]] )

for i in range(num_pair, num_pair+2*num_prim):
    for j in range(num_pair):
        S_twist[i,j] = better_simplify( S[twist_index[i][0],twist_index[j][0]]*S[twist_index[i][0],twist_index[j][1]] )
    for j in range(num_pair,num_pair+2*num_prim):
        S_twist[i,j] = better_simplify( S[twist_index[i][0],twist_index[j][0]]**2/2 )
    for j in range(num_pair+2*num_prim,num_pair+4*num_prim):
        S_twist[i,j] = better_simplify( ind_sign[twist_index[i][1]]*S[twist_index[i][0],twist_index[j][0]]/2 )

for i in range(num_pair+2*num_prim,num_pair+4*num_prim):
    for j in range(num_pair,num_pair+2*num_prim):
        S_twist[i,j] = better_simplify( ind_sign[twist_index[j][1]]*S[twist_index[i][0],twist_index[j][0]]/2 )
    for j in range(num_pair+2*num_prim,num_pair+4*num_prim):
        S_twist[i,j] = mathematica( ind_sign[twist_index[j][1]]*ind_sign[twist_index[i][1]]*Comb_Mat[twist_index[i][0],twist_index[j][0]]\
        *exp(I*pi*(cdim[twist_index[i][0]]+cdim[twist_index[j][0]]-c/12)) / 2 ).Simplify().sage()

# Now we define the modular T-matrix for the twisted characters
T_twist = diagonal_matrix( SR, [ exp(2*pi*I*(wght-c/12)) for wght in weight_index ] )


'''
We now turn to the coset model
'''
# We first compute the conformal weights in the coset model
# Define some compile time parameters
DEF k = 8 # level
DEF num_prim_cos = k*(k+1) // 2 # number of primary fields (primary with respect to the chiral algebra, of course)

# Define a fundamental domain for l,m to live in; our convention is that l,m are integers (twice the physics convention).
cdef int m_max(int l):
    if l <= k/2:
        return l
    else:
        return l-2
cdef int l,m,n
fund = [(l,m) for l in range(k+1) for m in range(-m_max(l), m_max(l)+1, 2)]
# This is just a weird degeneracy not covered by the above
if k % 2 == 0:
    fund.remove( (k//2, -k//2) )

# I don't think this can ever happen, but if it does something went very wrong.
if num_prim_cos != len(fund):
    raise RuntimeError( 'The number of indices in the fundamental domain does not match the number of primaries.' )

# The conformal weights
cdef list coset_dim = [ QQ(l*(l+2)) / QQ(4*(k+2)) - QQ(m**2) / QQ(4*k) for (l,m) in fund ]

# Dictionary to store the conformal weights the coset representation can decompose into
cdef dict dposs_decomp = {ccw:set() for ccw in coset_dim}
for ccw in coset_dim:
    for cw in conf_w:
        if QQ(ccw - cw).denominator() == 1:
            dposs_decomp[ccw].add(cw)
            

def chi_int_cos(int n_max, int l, int m):
    '''
    This computes the charcters in the coset model (but only the integer powers; the q^{h-c/24} must be added afterwards)
    '''
    cdef int r,s,ind1,ind2,sgn,eff_n_max = 2*n_max
    P = PowerSeriesRing(ZZ, 'q', sparse = False, default_prec = eff_n_max)
    cdef list series = [0] * ( eff_n_max +1 )
    for r in range(eff_n_max):
        for s in range(eff_n_max):
            ind1 = ( r*(r+1+l+m) + s*(s+1+l-m) )//2 + r*s*(k+1)
            ind2 = ( r*(r+1-l-m) + s*(s+1-l+m) )//2 + (k+1)*(r+s+1+r*s) - l
            sgn = 1 if (r+s)%2 == 0 else -1
            if ind1 <= eff_n_max:
                series[ind1] += sgn
            if ind2 <= eff_n_max:
                series[ind2] -= sgn
    return P(series, prec=eff_n_max)*qexp_eta(P, eff_n_max)**(-2)

'''
Now we compute the decompositions
We will loop through the fundamental domain fund
At each step we will compute the coset model character
Then we will compute it's decomposition
'''

cdef list tri_chars,expo
solutions = [] # This is not a cdef so it will be accessible outside this cell
cdef int num_char
for (l,m) in fund:
    weight = QQ(l*(l+2)) / QQ(4*(k+2)) - QQ(m**2) / QQ(4*k)
    char = P(q**(weight-c/12),prec=n_max)*P(chi_int_cos(n_max, l, m))
    tri_chars = []
    poss_labels = []
    for w in dposs_decomp[weight]:
        tri_chars += list( twist_char[w].values() )
        poss_labels += [ (w,key) for key in twist_char[w] ]
    num_char = len(tri_chars)

    for n in range(num_char+1,n_max):
        expo = char.exponents()[:n]
        V = vector(ZZ, n)
        M = matrix(ZZ, num_char, n)
        for i in range(n):
            V[i] = char[expo[i]]
            for j in range(num_char):
                M[j,i] = tri_chars[j][expo[i]]
        # If the equation has no solution a ValueError is raised and we continue
        try:
            this_sol = M.solve_left(V)
        except ValueError:
            continue
        this_kern = M.kernel().basis()
        com_char = sum(P(this_sol[i]*tri_chars[i],prec = n_max) for i in range(num_char))
        if len(this_kern)==0 or n == n_max-1:
            if P(com_char-char, prec = n_max-1) == P(0, prec = n_max-1):
                solutions += [[QQ(l*(l+2)) / QQ(4*(k+2)) - QQ(m**2) / QQ(4*k), (l,m), this_sol, poss_labels]]
            break

# Now we want to compute the M-matrix for gauging by the Z_2 defect eta on the torus
eta_tri = [1,1,1,1,-1,-1] # The action of eta on tricritcal ising

# The four different sub diagrams
M_twist_id_id = identity_matrix(SR, num_pair+4*num_prim)
#M_twist_id_eta = diagonal_matrix(SR, [eta_tri[i] * eta_tri[j] for i in range(num_prim) for j in range(i)]+[1 for _ in range(num_prim)]*4)
M_twist_id_eta = diagonal_matrix(SR, [eta_tri[i] * eta_tri[j] for i in range(num_prim) for j in range(i)]+[1]*2*num_prim+eta_tri*2)
M_twist_eta_id = ( S_twist.H * M_twist_id_eta * S_twist ).simplify()
M_twist_eta_eta = ( T_twist.H * M_twist_eta_id * T_twist ).simplify()

# The M_matrix
M_twist = ( ( M_twist_id_id + M_twist_id_eta + M_twist_eta_id + M_twist_eta_eta ) / 2 ).simplify()

# Set of modules that appear in the gauged theory. These are given by non-zero entries in M
modules = set()

# Simplify the M-matrix (this method is significantly faster than just applying FullSimplify to the whole thing)
for i in range(M_twist.nrows()):
    for j in range(M_twist.ncols()):
        if abs( M_twist[i,j].n() ) > 1e-4:
            M_twist[i,j] = better_simplify( M_twist[i,j] )
            modules.add(twist_index[i])
            modules.add(twist_index[j])
        else:
            M_twist[i,j] = 0


def Zse_dir():
    '''
    Function to compute the Parition function gauged by the exchange and diagonal Z_2 Verlinde line
    It will compute it up to the order n_max
    '''
    Z_out = P(0)
    cdef int i,j
    for i in range(num_pair+4*num_prim):
        for j in range(num_pair+4*num_prim):
            if M_twist[i,j] != 0:
                Z_out +=  P( M_twist[i,j] ) * P( twist_char[index_to_weight[twist_index[i]]][twist_index[i]] )\
                *P( twist_char[index_to_weight[twist_index[j]]][twist_index[j]] )
    return Z_out
```

# Gauging Tricritical Ising squared with discrete torsion.

```python
%%time
import itertools as itr
'''
This code computes a Z_2\times Z_2 group cocycle called dsign.
Then it computes the gauged tricritical ising squared partition function with discrete torsion given by dsign
There is a loop over all such discrete torsions with the resulting partition functions stored in Z_torsion
'''

# Set to store all partition functions with torsion
Z_torsion = []

# Some gap code to generate a representative of all cohomology classes in H^2(Z_2^2,Z_2) \cong Z_2^3
gap( 'LoadPackage("hap");\
    G:=Group((1,2),(3,4));;\
    A:=Group((1,2));;\
    A:=TrivialGModuleAsGOuterGroup(G,A);;\
    R:=ResolutionFiniteGroup(G,4);;\
    C:=HomToGModule(R,A);;\
    CH:=CohomologyModule(C,2);;\
    classes:=Elements(ActedGroup(CH))' )

# Group generated by sigma (exchange defect) and eta (product of the two Z_2 Verlinde lines in each copy of Tricritical Ising)
# This is a SageMath permutation group Z_2^2
our_G = CyclicPermutationGroup(2).direct_product( CyclicPermutationGroup(2), False ) 

# The group elements as GAP objects
e = gap( Permutation(srange(1,5)) )
eta = gap( our_G.gen(0) )
sig = gap( our_G.gen(1) )
es = eta*sig
# A list of all the group elements
group = [ e, eta, sig, es ]

'''
This is a loop over the possible group 2-cocycles indexed by the variable cocycle_index;
cocycle_index is a number between 1 and 8 indicating which cocycle should be computed.
There are only two distinct cohomology classes in H^2(Z_2^2,U(1)), but there is a surjective 4->1 mapping from H^2(Z_2^2,Z_2).
'''
num_each_class = []
for cocycle_index in range(1,9): # Index goes from 1 to 9 because GAP indexes from 1
    # The following GAP code generates a representative of the group cohomology class corresponding to 'cocycle_index'
    f = gap( f'Mapping( CH!.representativeCocycle(classes[{cocycle_index}]) )' )

    # Function that takes in two GAP group elements a,b and outputs a sign corresponding to the value of f(a,b) / f(b,a)
    def dsign(a,b):
        if f(a,b) == f(b,a):
            return 1
        elif f(a,b) == gap( Permutation([2,1]) ):
            return -1
        elif f(b,a) == gap( Permutation([2,1]) ):
            return -1
        else:
            raise ValueError('The GAP computed cocycle does not return an elemant in Z_2.')

    # Actually computes the partition function under discrete torsion
    Zsd = Z(100, [[Mpp, 1, 1, dsign(e,e), 2], [Mpm, 1, 1, dsign(e,eta), 2], [Mmp, 1, 1, dsign(eta,e), 2], [Mmm, 1, 1, dsign(eta,eta), 2],\
    [Mmp,1,2, dsign(eta,es)+dsign(eta,sig), 1], [Mpm,1,1/2, dsign(es,eta)+dsign(sig,eta), 1], [Mpm,-1,1/2, dsign(es,sig)+dsign(sig,es), 1],\
    [Mpp, 1, 2, dsign(e,sig)+dsign(e,es), 1], [Mpp, 1, 1/2, dsign(sig,e)+dsign(es,e), 1], [Mpp, -1, 1/2, dsign(sig,sig)+dsign(es,es), 1]])/4

    # Add to the list of all partition functions and print out the result
    if Zsd not in Z_torsion:
        Z_torsion += [Zsd]
    # There are only two distinct values possible, so once they are found, stop computing!
    if len(Z_torsion) >= 2:
        break
```

# Compute the characters and partition function of the coset model $(\mathfrak{su}(2)_M \oplus \mathfrak{su}(2)_N) / \mathfrak{su}(2)_{M+N}$ with $M=8$ and $N=2$.
This theory should be the same as tricritical Ising gauged by the exchange and diagonal $\mathbb{Z}_2$ Verlinde line with non-trivial discrete torsion when the AAE modular invariant is chosen. 

```python
%%cython
from sage.rings.power_series_ring import PowerSeriesRing
from sage.rings.puiseux_series_ring import PuiseuxSeriesRing
from sage.rings.rational_field import QQ
from sage.rings.integer_ring import ZZ
from sage.symbolic.ring import SR

from sage.functions.other import ceil
from sage.misc.functional import symbolic_prod as product

from sage.matrix.constructor import matrix
from sage.matrix.special import identity_matrix

from sage.symbolic.constants import pi
from sage.functions.trig import sin
from sage.misc.functional import sqrt
from sage.modular.etaproducts import qexp_eta 

import itertools as itr

from sage.interfaces.mathematica import mathematica

'''
This block calculates the partition function of ( su(2)_M \oplus su(2)_N ) / su(2)_{N+M}.
You can specify any valid modular invariant.
Isn't that cool!
'''

# The levels of the two SU(2)_{} factors in the numerator
DEF M = 8
DEF N = 2

DEF p = M+N+2
DEF pp = M+2

# Some parameters of the coset model
DEF dim_su2 = 3
c_charge = QQ(dim_su2)*( QQ(M)/QQ(M+2) + QQ(N)/QQ(N+2) - QQ(M+N)/QQ(M+N+2) ) # central charge

# Specify the modular invariants
# First entry corresponds to su(2)_M, second to su(2)_N, and third to su(2)_{M+N}
# Must be one of A,D,E, but the subscript on the series will be inferred
# If a non-existent modular invariant is given an error is raised
# See the BYB (17.114) on page 742 for a complete classification and details
modular_invariant = [ 'D', 'A', 'E' ]

# Get the mass matrices of the three su(2)_{}
mass_matrices = []
levels = [ZZ(M), ZZ(N), ZZ(M+N)]
for i in range(3):
    ADE = modular_invariant[i]
    if ADE == 'A': # A_{level+1}
        mass_matrices += [identity_matrix(SR, levels[i]+1)]
    elif ADE == 'D':
        mass_matrix = matrix(SR, levels[i]+1)
        if levels[i]%4==0: # D_{2l+2}
            l = levels[i] // 4
            mass_matrix[2*l,2*l] = 2
            for j in range(0, 2*l-1, 2):
                mass_matrix[j,j] = 1
                mass_matrix[j,4*l-j] = 1
                mass_matrix[4*l-j,j] = 1
                mass_matrix[4*l-j,4*l-j] = 1
        elif levels[i]%4==2: # D_{2l+1}
            l = (levels[i]+2) // 4
            mass_matrix[2*l-1,2*l-1] = 1
            for j in range(0, 4*l-1, 2):
                mass_matrix[j,j] = 1
            for j in range(1, 2*l-2, 2):
                mass_matrix[j,4*l-2-j] = 1
                mass_matrix[4*l-2-j,j] = 1
        else:
            raise ValueError('Can only specify \'D\' series for even levels.')
        mass_matrices += [mass_matrix]
    elif ADE == 'E':
        if levels[i] == 10: # E_6
            mass_matrix = matrix(SR, 11)
            for i,j in [(0,6), (3,7), (4,10)]:
                mass_matrix[i,i] = 1
                mass_matrix[j,j] = 1
                mass_matrix[i,j] = 1
                mass_matrix[j,i] = 1
        elif levels[i] == 16: # E_7
            mass_matrix = matrix(SR, 17)
            for i,j in [(0,16), (4,12), (6,10)]:
                mass_matrix[i,i] = 1
                mass_matrix[j,j] = 1
                mass_matrix[i,j] = 1
                mass_matrix[j,i] = 1
            mass_matrix[8,8] = 1
            mass_matrix[8,2] = 1
            mass_matrix[2,8] = 1
            mass_matrix[8,14] = 1
            mass_matrix[14,8] = 1
        elif levels[i] == 28: # E_8
            mass_matrix = matrix(SR, 29)
            for i,j in itr.product([0,10,18,28],repeat=2):
                mass_matrix[i,j] = 1
            for i,j in itr.product([6,12,16,22], repeat=2):
                mass_matrix[i,j] = 1
        else:
            raise ValueError('Can only specify \'E\' series for levels equal to 10, 16, and 28.')
        mass_matrices += [mass_matrix]
    else:
        raise ValueError('Modular invariant must be one of \'A\', \'D\', or \'E\'.')

# Lets now generate a fundamental domain
# The fixed point at (a,b,c) = (k/2,l/2,(k+l)/2) is only one primary
dcoset_fund = []
not_fund = set()
for a,b,c in itr.product(range(M+1), range(N+1), range(M+N+1)):
    if (a+b-c)%2 == 0 and (a,b,c) not in not_fund: # Spin selection rules and field identifications
        dcoset_fund.append( (a,b,c) )
        not_fund.add( (M-a,N-b,M+N-c) )

fixed_points = set(dcoset_fund).intersection(not_fund)
not_fund = list(not_fund)
dcoset = dcoset_fund+not_fund

# The number of primaries in the coset
cdef int num_prim_cos = len(dcoset_fund)

# Calculate the modular invariant of the coset model using the fundamental domain
# and the selected ADE matrices above.
M_dcoset = matrix(SR, num_prim_cos)
for i,j in itr.product(range(num_prim_cos), repeat=2):
    ind_i = (dcoset[i], (M-dcoset[i][0], N-dcoset[i][1], N+M-dcoset[i][2]) )
    ind_j = (dcoset[j], (M-dcoset[j][0], N-dcoset[j][1], N+M-dcoset[j][2]) )
    M_dcoset[i,j] = sum(product(mass_matrices[l][ind_i[a][l],ind_j[b][l]] for l in range(3)) for a,b in itr.product(range(2),repeat=2))/SR(2)


# Now calculate the modular S-matrix for the coset
# First calculate the S-matrices for SU(2)_M, SU(2)_N and SU(2)_{N+M}
def su2_S_mat(int level):
    cdef list Smat = [ [0 for _ in range(level+1)] for _ in range(level+1) ]
    for i,j in itr.product(range(level+1),repeat=2):
        Smat[i][j] = sqrt( SR(2) / SR(level+2) ) * sin( pi*SR((i+1)*(j+1)) / SR(level+2) )
    return matrix(SR, Smat)
    
# Now compute the diagonal coset S-matrix
su2_S_matrices = [ su2_S_mat(M), su2_S_mat(N), su2_S_mat(M+N) ]
S_dcoset = matrix(SR, num_prim_cos)
for i,j in itr.product(range(num_prim_cos), repeat=2):
    S_dcoset[i,j] = mathematica( SR(2)*product( su2_S_matrices[l][dcoset_fund[i][l],dcoset_fund[j][l]] for l in range(3) ) )\
                    .Simplify().sage()

# Here's a function that computes the su(2)_{level} string functions
def string_function_int(int level, int l, int m, int max_ord):
    '''
    Function to compute the level 'level' string function c_n^l up to an order given by max_ord.
    Only integer powers are given, multiply by q^s, s(m,l,level) = l(l+2)/(4(level+2)-m^2/(4level)-level/(8(level+2)),
    to get the actual function
    '''
    cdef int max_r,max_s,r,s,ind1,ind2,sgn
    # max_r and max_s could be set slightly lower if optimising this code became really critical
    # As it is it runs in a coupke milliseconds for max_ord = 1000, so it's not really an issue
    max_r = max_ord
    max_s = max_ord
    P = PowerSeriesRing(ZZ, 'q', sparse = False, default_prec = max_ord)
    cdef list series = [0] * ( max_ord +1 )
    for r in range(max_r):
        for s in range(max_s):
            ind1 = ( r*(r+1+l+m) + s*(s+1+l-m) )//2 + r*s*(level+1)
            ind2 = ( r*(r+1-l-m) + s*(s+1-l+m) )//2 + (level+1)*(r+s+1+r*s) - l
            sgn = 1 if (r+s)%2 == 0 else -1
            if ind1 <= max_ord:
                series[ind1] += sgn
            if ind2 <= max_ord:
                series[ind2] -= sgn
    return P(series, prec=max_ord)*qexp_eta(P, max_ord)**(-3)


# This function will compute the coset characters
def coset_char(int r, int l, int s, list strings, int max_ord):
    '''
    This function computes the coset characters of ( su(2)_M \oplus su(2)_N ) / su(2)_{N+M}.
    r-1, l, s-1 are the su(2))_M, su(2)_N, and su(2)_{M+N} reps, respectively
    strings is an N by N list of the su(2)_N string functions
    max_ord is the maximum order of q to compute the q-expansion of the character up to
    '''
    P = PuiseuxSeriesRing(ZZ, 'q', sparse = True, default_prec = max_ord)
    q = P.gen()
    if (r+l+s)%2!=0: # Know this by selection rules
        return P(0,prec=max_ord)
    if not ( 0<r<pp and 0<=l<=N and 0<s<p ):
        raise ValueError(f'No representation corresponds to the triple {{{r},{l};{s}}}')
        
    char = P(0,prec=max_ord)

    cdef list str_mult
    cdef int e_char = 4*N*p*pp
    cdef j,n,ind1,ind2,max_ord_times_e = max_ord*e_char
    
    for n in range(-N+1, N+1):
        if (n+l)%2:
            continue
        str_mult = [0] * (max_ord_times_e + 1)
        for j in range(-max_ord*2*p, max_ord*2*p, 2*p):
            if not (s+j-r-n)%(2*N):
                ind1 = ((s+j-r)*pp-r*N)**2
                if ind1 <= max_ord_times_e:
                    str_mult[ ind1 ] += 1
            if not (-s+j-r-n)%(2*N):
                ind2 = ((-s+j-r)*pp-r*N)**2
                if ind2 <= max_ord_times_e:
                    str_mult[ ind2 ] -= 1

        char += ( P( strings[l][abs(n)], prec=max_ord, e=1 ) * P( str_mult, prec=max_ord, e=e_char ) ) * \
                q**( QQ(l*(l+2))/QQ(4*(N+2))-QQ(n**2)/QQ(4*N)-QQ(N)/QQ(8*(N+2)) )
        
    return char


def partition_function(int max_ord):
    '''
    Computes the partition function of the coset SU(2)_M\times SU(2)_N / SU(2)_{M+N} up to order max_ord
    '''
    
    # The power series ring
    P = PuiseuxSeriesRing(ZZ, 'q', sparse = True, default_prec = max_ord)
    # Pre-compute all the string functions
    # There are redundancies here, but this computation is not a performance bottleneck, so we won't worry about them!
    cdef list strings = [[string_function_int(N, l, n, max(15, max_ord)) for n in range(N+1)] for l in range(N+1)]
    # Now we loop to compute the partition function
    cdef list characters = []
    pfunc = P(0, prec=max_ord)
    for i in range(num_prim_cos):
        (lam,mu,nu) = dcoset_fund[i]
        characters += [ P( coset_char(lam+1, mu, nu+1, strings, max_ord) ) ]
        if M_dcoset[i,i] != 0:
            pfunc += P( M_dcoset[i,i] )*characters[i]*characters[i]
        for j in range(i):
            if M_dcoset[i,j] != 0:
                pfunc += P( M_dcoset[i,j]+M_dcoset[j,i] )*characters[i]*characters[j]
    return pfunc
```

```python
%%cython
from sage.rings.power_series_ring import PowerSeriesRing
from sage.rings.puiseux_series_ring import PuiseuxSeriesRing
from sage.rings.rational_field import QQ
from sage.rings.integer_ring import ZZ

from sage.matrix.constructor import matrix
from sage.modules.free_module_element import vector

from sage.misc.functional import isqrt

from sage.combinat.partition import number_of_partitions
import itertools as itr

from __main__ import coset_char, string_function_int, chi_int, dcoset_fund, M_dcoset

'''
This block defines a function that takes two sets of characters and tries decomposes one into the other
It then does this decomposition for the diagonal coset characters being decomposed into the characters of the invariant algebra
'''

# Some global parameters
DEF N = 2
DEF M = 8
DEF max_ord = 100

def char_decomp(list char1, list char2, int max_ord):
    '''
    Function to compute the decomposition of the characters in char1 into the characters in char2.
    max_ord is the order up to which the decomposition is checked
    A list of lists called solutions is returned.
    The number solutions[i][j] is the number of times character j in char2 appears in the decomposition of character i in char1
    '''

    # Power Series Ring the characters live in
    P = PuiseuxSeriesRing(QQ, 'q', sparse=True, default_prec=max_ord)

    # To speed things up, let us first check which decompositions are in principle possible 
    # according to the fractional part of the character exponents
    cdef list poss_decomps = [[] for _ in range(len(char1))]
    cdef int i,j
    cdef list val1 = [ QQ( char1[i].valuation() ) for i in range(len(char1)) ]
    cdef list val2 = [ QQ( char2[i].valuation() ) for i in range(len(char2)) ]
    for i,j in itr.product(range(len(char1)), range(len(char2))):
        if (val1[i]-val2[j]).denominator() == 1:
            poss_decomps[i] += [j]
    print(poss_decomps)
            
    # Now we actually compute the decompositions
    cdef list expo
    cdef int num_char,max_ord_eff,n
    cdef int l,k
    # List of solutions
    cdef list solutions = [[] for _ in range(len(char1))]
    max_num = min([len(char.exponents())-1 for char in char2] + [len(char.exponents())-1 for char in char1])
    print(max_num)
    for i in range(len(char1)):
        # To save time just skip ones where no decomposition is possible.
        if not poss_decomps[i]:
            continue
        # We now try solving it up to order 'n' and increase n until we fix a unique solution
        num_char = len(poss_decomps[i])
        char = char1[i]
        for n in range(num_char, max_num):
            expo = char.exponents()[:n]
            Vec = vector(ZZ, n)
            Mat = matrix(ZZ, num_char, n)
            for l in range(n):
                Vec[l] = char[expo[l]]
                for k in range(num_char):
                    Mat[k,l] = char2[poss_decomps[i][k]][expo[l]]
            # If the equation has no solution a ValueError is raised and we continue
            try:
                this_sol = Mat.solve_left(Vec)
            except ValueError:
                continue
            this_kern = Mat.kernel().basis()
            
            if len(this_kern)==0 or n == max_num-1:
                com_char = sum(P(this_sol[j]*char2[poss_decomps[i][j]], prec = max_ord) for j in range(num_char))
                if P(com_char-char, prec = max_ord-1) == 0:
                    solutions[i] += [ this_sol, poss_decomps[i] ]
                break

    # Return all the decompositions we have found
    return solutions

'''
The following code will compute the characters with respect to the invariant algebra
'''
# Minimal Model Fixed Parameters

# This will be the (p,p'=pp) minimal model
DEF p = 5  # p
DEF pp = 4  # p'
# Various constant parameters of the minimal model
DEF num_prim = (p-1)*(pp-1) // 2  # Number of primaries

c = QQ(1) - QQ(6 *(p-pp)**2)/QQ(p * pp) # Central charge

# Index set
####### This needs to be changed for each minimal model ########
Epp = [ (1,1), (1,2), (1,3), (1,4), (2,2), (2,4) ] # By (r,s)
# The conformal weights
cdim = [ (QQ(p*r-pp*s)**2 - QQ(p-pp)**2) / QQ(4*p*pp) for r,s in Epp ]

# For faster computation a power series ring with only integer powers is defined
PZ = PowerSeriesRing(ZZ, 'q', sparse = False, default_prec = max_ord)

# Power Series Ring the characters will live in
P = PuiseuxSeriesRing(QQ, 'q', sparse = True, default_prec = max_ord)
q = P.gen()


def shift_series(ser, shift, prec = max_ord):
    '''
    Helper function to shift the powers of a series by some fixed amount called 'shift'
    '''
    
    nshift = shift.numerator()
    dshift = shift.denominator()
    try:
        ram = QQ(ser.ramification_index())
    except AttributeError:
        return P({ZZ(k*dshift + nshift) : ser[k] for k in ser.exponents()}, dshift, prec)
    else:
        return P({ZZ(ram*rat*dshift+ram*nshift) : ser[rat] for rat in ser.exponents()}, dshift*ram, prec)

# This code actually computes the characters with respect to the invariant algebra
# These include the characters twisted by the exchange defect hence the name 'twist_char' for the dictionary storing them
# Because characters with respect to the invariant algebra is a mouthfull, 
# I abuse terminology and call all these characters twist characters in the comments
cdef int i,j,k
cdef list all_chi = [ 0 for _ in range(num_prim) ]
twist_char = {}
for i in range(num_prim):
    all_chi[i] = PZ( chi_int(2*max_ord+1, Epp[i][0], Epp[i][1]), prec=2*max_ord+1 )
    this_chi = PZ( all_chi[i] )

    # 'Diagonal' untwisted twist characters
    twist_char[(i,'p')] = shift_series( ( this_chi**2 + PZ({2*k:this_chi[k] for k in range(max_ord)}) )/2, 2*cdim[i]-c/12, max_ord )
    twist_char[(i,'m')] = shift_series(  ( this_chi**2 - PZ({2*k:this_chi[k] for k in range(max_ord)}) )/2, 2*cdim[i]-c/12, max_ord )
    # Twisted twist characters
    twist_char[(i,'tp')] = shift_series( P([this_chi[k]*(1-k%2) for k in range(2*max_ord)], QQ(2)), cdim[i]/2-c/48, max_ord )
    twist_char[(i,'tm')] = shift_series( P([this_chi[k]*(k%2) for k in range(2*max_ord)], QQ(2)), cdim[i]/2-c/48, max_ord )

    # Untwisted twist characters 
    for j in range(i):
        twist_char[(i,j)] = shift_series(this_chi*PZ(all_chi[j]), cdim[i]+cdim[j]-c/12, max_ord)

# Now we define an index list for the twisted characters to go between the (i,something) notation and a single index
twist_index = [ (i,j) for i in range(num_prim) for j in range(i) ] + [ (i,'p') for i in range(num_prim) ]\
+ [ (i,'m') for i in range(num_prim) ] + [ (i,'tp') for i in range(num_prim) ] + [ (i,'tm') for i in range(num_prim) ]
# Number of twisted characters
cdef int num_twist_char = len( twist_index )


'''
Now we organise the two lists of characters char1 and char2 and actually compute the decompositions!!
'''
cdef list strings = [ [string_function_int(N, l, n, max(10, max_ord)) for n in range(N+1)] for l in range(N+1) ]
char1 = []
char_ind = []
for r,l,s in dcoset_fund:
    char_ind += [(r,l,s)]
    char1 += [ coset_char(r+1, l, s+1, strings, max_ord) ]

# Twist characters
#char2 = [twist_char[twist_index[i]] for i in range(num_twist_char)]

# Tricritical Ising Squared Characters
char2 = [ shift_series(PZ(all_chi[i]*all_chi[j]), cdim[i]+cdim[j]-c/12, max_ord) for i in range(num_prim) for j in range(i+1) ]


decompositions = char_decomp(char1, char2, max_ord-10)
```
